## Split Data into Training and Testing Sets

### Subtask:
Divide the preprocessed dataset into training and testing sets (e.g., 80% for training, 20% for testing) to prepare for model training and evaluation.


**Reasoning**:
I need to import the `train_test_split` function from `sklearn.model_selection` and then use it to split the `X` and `Y` data into training and testing sets as per the instructions, setting `test_size=0.2` and a `random_state` for reproducibility.

## Visualize Relationships with Target

### Subtask:
Generate visualizations, such as count plots or bar charts, to show the distribution of key categorical features (e.g., 'gender', 'ethnicity', 'smoking_status') in relation to the 'diagnosed_diabetes' target variable. Include appropriate labels and a legend.


**Reasoning**:
I need to import `matplotlib.pyplot` and `seaborn` to create the requested count plots. These libraries are essential for data visualization in Python.



```python
import matplotlib.pyplot as plt
import seaborn as sns

print("Matplotlib and Seaborn imported successfully.")
```

**Reasoning**:
I need to generate count plots for the 'gender', 'ethnicity', and 'smoking_status' columns, showing their distribution relative to the 'diagnosed_diabetes' target variable, as specified in the instructions. This will involve using `seaborn.countplot` and `matplotlib.pyplot` for visualization.



```python
plt.figure(figsize=(15, 5))

# Plot for gender
plt.subplot(1, 3, 1)
sns.countplot(data=df, x='encoded_gender', hue='diagnosed_diabetes')
plt.title('Distribution of Diabetes Diagnosis by Gender')
plt.xlabel('Encoded Gender')
plt.ylabel('Count')
plt.legend(title='Diagnosed Diabetes')

# Plot for ethnicity
plt.subplot(1, 3, 2)
sns.countplot(data=df, x='encoded_ethnicity', hue='diagnosed_diabetes')
plt.title('Distribution of Diabetes Diagnosis by Ethnicity')
plt.xlabel('Encoded Ethnicity')
plt.ylabel('Count')
plt.legend(title='Diagnosed Diabetes')

# Plot for smoking_status
plt.subplot(1, 3, 3)
sns.countplot(data=df, x='encoded_smoking_status', hue='diagnosed_diabetes')
plt.title('Distribution of Diabetes Diagnosis by Smoking Status')
plt.xlabel('Encoded Smoking Status')
plt.ylabel('Count')
plt.legend(title='Diagnosed Diabetes')

plt.tight_layout()
plt.show()
```

## Train Classification Models

### Subtask:
Train three different classification models: K-Nearest Neighbors (KNeighborsClassifier), Decision Tree (DecisionTreeClassifier), and Support Vector Machine (SVC) using the training data (X_train, Y_train).
