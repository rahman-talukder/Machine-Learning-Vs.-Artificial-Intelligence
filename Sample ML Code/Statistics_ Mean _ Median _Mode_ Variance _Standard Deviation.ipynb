```python
import numpy as np
import statistics as stats

```

```python
# Mean (গড় / গড় মান)

# একটি ডেটাসেটের সব মান যোগ করে মানের সংখ্যায় ভাগ করলে যে মান পাওয়া যায় তাকে Mean বা গড় বলা হয়।
# উদাহরণ: (5 + 10 + 15) / 3 = 10
heights=[168,170,150,160,182,140,175,191,152,150]
mean=np.mean(heights)
print(mean)
```

```python
#-------------------------------------------------------
#Step->> 1: ডেটা ছোট থেকে বড় অনুযায়ী সাজালে মাঝের সংখ্যাটিকে Median বলে।(Median (মধ্যক))
#-------------------------------------------------------

Odd =[1, 2, 3, 4, 5]    #মানের সংখ্যা বিজোড় হলে → মাঝের সংখ্যা (এখানে median 3)
Even =[1, 2, 3, 4 ]     #মানের সংখ্যা জোড় হলে → মাঝের দুই সংখ্যার গড় (প্রথমে দুটি )
```

```python
# Sort  of Data increasing order
heights.sort()
heights
```

```python
# Median (মধ্যক)

# ডেটা ছোট থেকে বড় অনুযায়ী সাজালে মাঝের সংখ্যাটিকে Median বলে।
# মানের সংখ্যা বিজোড় হলে → মাঝের সংখ্যা
# মানের সংখ্যা জোড় হলে → মাঝের দুই সংখ্যার গড়

median=np.median(heights)
print(f'Median of Data: {median}')
```

```python
# Mode (প্রচুরক)

# ডেটাসেটে যে মানটি সবচেয়ে বেশি বার আসে, সেটি Mode।
# উদাহরণ: 2, 3, 3, 5 → Mode = 3
import statistics as stats
mode=stats.mode(heights)
print(f'Mode of Data: {mode}')
```

```python
# Variance (প্রসরণ)
# ডেটাসেটের মানগুলো গড় থেকে কতটা ছড়ানো তা বোঝায়।
# এটি প্রতিটি মান ও গড়ের পার্থক্যের বর্গ যোগ করে মোট মানের সংখ্যায় ভাগ করে হিসাব করা হয়।

variance=np.var(heights)
print(f'Variance of Data: {variance}[0]:.2f')
```

```python
পরিসংখ্যান বা স্ট্যাটিসটিক্সে এটি একটি অত্যন্ত গুরুত্বপূর্ণ ধারণা। সহজ কথায়, কোনো তথ্যের বা ডেটা সেটের মানগুলো তাদের গড় (Mean/Average) মান থেকে কতটা ছড়িয়ে ছিটিয়ে আছে, তা পরিমাপ করাই হলো পরিমিত ব্যবধান।

নিচে বিস্তারিত আলোচনা করা হলো:

১. মূল ধারণা (Basic Concept)

পরিমিত ব্যবধান বা Standard Deviation (SD) আমাদের বলে দেয় ডেটাগুলো গড়ের খুব কাছাকাছি আছে নাকি অনেক দূরে ছড়িয়ে আছে।

* **কম SD (Low SD):** এর মানে হলো ডেটাগুলো তাদের গড় মানের খুব কাছাকাছি অবস্থান করছে। (যেমন: একটি ক্লাসের সব ছাত্রের বয়স যদি ১১ বা ১২ হয়)।
* **বেশি SD (High SD):** এর মানে হলো ডেটাগুলো গড় মান থেকে অনেক দূরে বা বিক্ষিপ্ত অবস্থায় আছে। (যেমন: একটি ক্লাসের ছাত্রদের বয়স যদি ৫ থেকে ৫০ এর মধ্যে ছড়ানো থাকে)।

২. সূত্র (Formula)

পরিমিত ব্যবধান বের করার সাধারণ সূত্রটি হলো:

যেখানে:

*  (সিগমা) = পরিমিত ব্যবধান (Standard Deviation)
*  (সামেশন) = যোগফল
*  = প্রতিটি ডেটার মান
*  (মিউ) = ডেটা সেটের গড় (Mean)
*  = মোট ডেটার সংখ্যা


৩. একটি সহজ উদাহরণ (Example Calculation)

ধরি, আমাদের কাছে ৫ জন ছাত্রের প্রাপ্ত নম্বর আছে: **১০, ১২, ২৩, ২৩, ১৬, ২৩, ২১, ১৬** (সহজ করার জন্য আমরা ছোট একটি সেট নিই: **৪, ৮, ৬, ৫, ৩, ২, ৮, ৯, ২, ৫** — *আসুন আরও সহজ একটি হিসাব দেখি যাতে পূর্ণসংখ্যা আসে।*

ধরি ডেটা সেট:

ধাপ ১: গড় (Mean) বের করা



তাহলে, গড় () = ৫।

ধাপ ২: প্রতিটি সংখ্যা থেকে গড় বিয়োগ করে তার বর্গ (Square) করা


ধাপ ৩: বর্গের যোগফল বের করা


ধাপ ৪: ভেরিয়েন্স (Variance) বের করা
যোগফলকে মোট সংখ্যা () দিয়ে ভাগ করতে হবে।

ধাপ ৫: পরিমিত ব্যবধান (Standard Deviation) বের করা
ভেরিয়েন্সের বর্গমূল (Square root) করলেই আমরা SD পাবো।


ফলাফল: এই ডেটা সেটের Standard Deviation বা পরিমিত ব্যবধান হলো ২।



৪. কেন এটি গুরুত্বপূর্ণ?

শেয়ার বাজার:কোনো শেয়ারের দাম কতটা ঝুঁকিপূর্ণ তা বুঝতে এটি ব্যবহৃত হয়। (SD বেশি মানে ঝুঁকি বেশি)।
আবহাওয়া: তাপমাত্রার ওঠানামা বুঝতে সাহায্য করে।
গুণগত মান নিয়ন্ত্রণ (Quality Control):কারখানায় উৎপাদিত পণ্যের মাপ বা ওজন ঠিক আছে কি না তা যাচাই করতে এটি লাগে।

আপনি কি চান আমি এক্সেল (Excel) বা গুগল শিট ব্যবহার করে কীভাবে সহজে Standard Deviation বের করতে হয় তা দেখিয়ে দিই?
```
