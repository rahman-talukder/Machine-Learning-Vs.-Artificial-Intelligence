```python
import numpy as np
import statistics as stats

```

```python
# Mean (গড় / গড় মান)

# একটি ডেটাসেটের সব মান যোগ করে মানের সংখ্যায় ভাগ করলে যে মান পাওয়া যায় তাকে Mean বা গড় বলা হয়।
# উদাহরণ: (5 + 10 + 15) / 3 = 10
heights=[168,170,150,160,182,140,175,191,152,150]
mean=np.mean(heights)
print(mean)
```

```python
#-------------------------------------------------------
#Step->> 1: ডেটা ছোট থেকে বড় অনুযায়ী সাজালে মাঝের সংখ্যাটিকে Median বলে।(Median (মধ্যক))
#-------------------------------------------------------

Odd =[1, 2, 3, 4, 5]    #মানের সংখ্যা বিজোড় হলে → মাঝের সংখ্যা (এখানে median 3)
Even =[1, 2, 3, 4 ]     #মানের সংখ্যা জোড় হলে → মাঝের দুই সংখ্যার গড় (প্রথমে দুটি )
```

```python
# Sort  of Data increasing order
heights.sort()
heights
```

```python
# Median (মধ্যক)

# ডেটা ছোট থেকে বড় অনুযায়ী সাজালে মাঝের সংখ্যাটিকে Median বলে।
# মানের সংখ্যা বিজোড় হলে → মাঝের সংখ্যা
# মানের সংখ্যা জোড় হলে → মাঝের দুই সংখ্যার গড়

median=np.median(heights)
print(f'Median of Data: {median}')
```

```python
# Mode (প্রচুরক)

# ডেটাসেটে যে মানটি সবচেয়ে বেশি বার আসে, সেটি Mode।
# উদাহরণ: 2, 3, 3, 5 → Mode = 3
import statistics as stats
mode=stats.mode(heights)
print(f'Mode of Data: {mode}')
```

```python
# Variance (প্রসরণ)
# ডেটাসেটের মানগুলো গড় থেকে কতটা ছড়ানো তা বোঝায়।
# এটি প্রতিটি মান ও গড়ের পার্থক্যের বর্গ যোগ করে মোট মানের সংখ্যায় ভাগ করে হিসাব করা হয়।

variance=np.var(heights)
print(f'Variance of Data: {variance}[0]:.2f')
```
